```python
#STRIPPING RATIO CALCULATOR
#Built by:Carly Edward Quaicoe
#UMaT Mining Engineering -Year 1
#Date:May 2026

print("Welcome to the Stripping Ratio Calculator ")
print("Built by Carly Edward Quaicoe-UMaT Mining Engineering")
print("-=====================================-")

#STEP 2 -GET INPUTS FROM THE USER

print("Enter your mine data below:")
print("")
waste=float(input("Enter volume of WASTE material (m^3):"))
ore=float(input("Enter volume of ORE material (m^3):"))
#STEP 3-CALCULATE STRIPPING RATIOS

#Basic Stripping Ratio
BSR=waste/ore
#Bank Equivalent Stripping Ratio
BESR=waste/ore
#Print the results
print("")
print("=============== RESULTS =============")
print(f"Waste Volume:{waste}m^3")
print(f"Ore Volume:{ore}m^3")
print("")
print(f"Basic Stripping Ratio:{BSR:.2f}")
print(f"BESR:{BESR:.2f}")
print("")
#Intercept the result
if BSR<=5:
  print("STATUS:Economically VIABLE to mine")
elif BSR<=8:
  print("STATUS:BODERLINE-needs cost analysis")
else:
  print("STATUS:WARING-may not be economical")
```

```python
#STEP 4 -ISR AND MAXIMUM ALLOWABLE SR

ore_price=float(input("Enter Gold Price ($/tonne):"))
mining_cost=float(input("Enter Mining cost ($/tonne):"))
processing__cost=float(input("Enter Processing Cost ($/tonne):"))
#Maxium Allowable Stripping Ratio
MASR=(ore_price-processing__cost)/mining_cost
print("")
print("==============ADVANCE RESULTS=========")
print(f"Maxium Allowanle SR:{MASR:.2f}")
print("")

if BSR <=MASR:
  print("DECISION:Mine is PROFITABLE to operate")
else:
  print("DECISION:Mine is NOT profitable at current costs")
```
